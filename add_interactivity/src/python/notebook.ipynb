```python
import math

import ipywidgets
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
```

```python
def plot_fct(w=1):
    X = np.random.uniform(0, 5, size=100)
    y = 2 * X + w * np.random.normal(size=100)
    plt.scatter(X, y)
ipywidgets.interact(plot_fct, w=(0,10, 0.5))
```

```python
moons = make_moons(n_samples=100, noise=0.1)
X, y = moons[0], moons[1]

plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
def plot_moons(samples=200, noise=0):
    moons = make_moons(n_samples=samples, noise=noise)
    X, y = moons[0], moons[1]
    plt.scatter(X[:, 0], X[:, 1], c=y)

ipywidgets.interact(plot_moons, samples=[200, 500, 1000], noise=(0, 2, 0.025))
```

```python
def plot_sin(start=0, end=30, factor=1, grid=False, plot_cos=False):
    X = np.linspace(start, end, (end - start) * 10)
    y = np.sin(X) * factor
    plt.grid(grid)
    plt.plot(X, y)
    if plot_cos:
        y = np.cos(X)
        plt.plot(X, y)
        
ipywidgets.interact(plot_sin, start=(0, 10, 1), end=(20, 50, 1), factor=(0, 5, 0.1), grid=False)
```

```python
x_start = -5
x_end = 5
X = np.linspace(x_start, x_end, 1000)
y = 1 / (1 + np.exp(-X))

def plot_sigmoid(x_in=0):
    y_in = 1 / (1 + math.exp(-x_in))
    
    plt.plot(X, y)
    plt.scatter(x_in, y_in, c="r")
    
ipywidgets.interact(plot_sigmoid, x_in=(x_start, x_end, 0.1))
```

```python
def plot_hist(mu=0, sigma=1, n=100, bins=10, color="blue"):
    plt.xlim(-20, 20)
    X = np.random.normal(mu, sigma, n)
    plt.hist(X, bins=bins, color=color)
    
ipywidgets.interact(
    plot_hist, 
    mu=(-10, 10, 0.5), 
    sigma=(0, 10, 0.1), 
    n=(10, 1000, 1), 
    bins=(1, 100, 1),
    color=["red", "green", "blue"]
)
```
